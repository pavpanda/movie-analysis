```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1920, 1940, 1960, 1980, 2000, 2020]
y = [12.82, 18.31, 8.66, 3.11, 1.49, 1.00]
plt.style.use("ggplot")
plt.plot(x, y)
plt.title("Today's equivalent of $1 vs Time")
plt.ylabel("Today's equivalent of $1")
plt.xlabel("Time (year)")
plt.savefig("inflation.png", bbox_inches='tight')
```

```python
x = ["Avengers: Endgame", "Avatar", "Titanic", "Star Wars: The Force Awakens", "Avengers: Infinity War", "Jurassic World",
     "The Lion King", "The Avengers", "Furious 7", "Frozen II"]
y = [2797800564, 2790439000, 2194439542, 2068223624, 2048359754, 1671713208, 1656943394, 1518812988, 1516045911, 1450026933]
plt.style.use("ggplot")
plt.bar(x, y)
plt.xticks(rotation=75)
plt.title("Worldwide gross by Movie ")
plt.ylabel("Money earned (billions)")
plt.xlabel("Movie")
plt.savefig("gross.png", bbox_inches='tight')
```
